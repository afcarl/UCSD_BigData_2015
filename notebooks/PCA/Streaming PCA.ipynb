```python
import sys
import numpy as np
from numpy import linalg as LA

import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import display, clear_output

from time import time,sleep
import math
import sklearn
```

```python
# Generate data

d=2    # number of dimensions
k=2    # number of vectors to combine = the dimension of the 
n=1000 # number of examples

#V=np.random.normal(size=(k,d))
V=np.array([[2,2],[-1,1]])
Random=np.random.normal(size=(n,k))
shift=np.array([2,-1])
X=np.dot(Random,V)+shift
np.shape(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
std=[math.sqrt(x) for x in pca.explained_variance_]
print 'Standard Deviations=',std
print 'fraction of Variance Explained=',pca.explained_variance_ratio_
Mean=pca.mean_
P=pca.components_
print 'Mean=',Mean
print 'Eigen-Vecs=',P
```

```python
plt.figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')
ax = plt.axes()
ax.scatter(X[:,0],X[:,1])
ax.scatter(Mean[0], Mean[1], c='r')
scale=1
for j in range(2):
    ax.arrow(Mean[0],Mean[1], scale*std[j]*P[j][0], scale*std[j]*P[j][1], head_width=0.3, head_length=0.3, fc='k', ec='r')
ax.grid()
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.show()
```

```python
### Averaging vectors
class run_aver:
    ## Initialize the object
    def __init__(self,alpha=0.0):
        self.t = 0.0
        self.s = None
        self.alpha = alpha

    def send(self,x):
        if self.t==0:
            self.s=x
        elif self.alpha==0:
            self.s=(x + self.t*self.s)/(self.t+1.0)
        else:
            self.s=self.alpha*x + (1-self.alpha)*self.s                           
        self.t+=1.0
        residual=x-self.s
        return residual
    
    def get_state(self):
        return self.s

    def get_sqrt(self):
        return math.sqrt(self.s)  # only for the case that the dimension is 1
```

```python
Av=[run_aver(alpha=0) for i in range(2)]
err=[run_aver(alpha=0) for i in range(2)]
sz=np.shape(X)[0]
L=[np.zeros(sz) for i in range(3)]

for i in range(sz):
    x=X[i,:]
    err[0].send(np.dot(x,x))
    L[0][i]=err[0].get_state()
    r=Av[0].send(x)
    err[1].send(np.dot(r,r))
    L[1][i]=err[1].get_state()
```

[oja's rule](http://en.wikipedia.org/wiki/Oja%27s_rule)

```python
plt.plot(L[0],label='original')
plt.plot(L[1],label='residual')
#plt.ylim([0,200])
#plt.xlim([0,200])

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.grid()
```

```python
### Finding Eigen-Vectors
class run_Eigen:
    ## Initialize the object
    def __init__(self):
        self.t = 0.0
        self.var = run_aver(alpha=0.01) # for estimating the 

    def send(self,x):
        if self.t==0:
            self.s=np.random.normal(x)
        else:
            s=self.s
            s=s+(np.dot(s,x)/math.sqrt(self.t))*x
            Norm=LA.norm(s)
            if Norm>0: 
                s=s/Norm
            else:
                s=np.random.normal(x) 
            self.s=s
        self.t+=1.0
        
        dot=np.dot(self.s,x)
        self.var.send(dot*dot)
        residual=x-dot*self.s
        return residual
    
    def get_state(self):
        return self.s*math.sqrt(self.var.get_state())
    
    def get_var(self):
        return self.var.get_state()
    
```

```python
kk=2 # number of eigen-vectors to search for.

# The modeling pipeline
Av=run_aver(alpha=0)  # find the mean
Eig=[run_Eigen() for i in range(kk)] # find the first kk eigen-vectors

# Compute the errors 
err=[run_aver(alpha=0) for i in range(kk+2)] # compute the average length of the residual.
sz=np.shape(X)[0]
L=[np.zeros(sz) for i in range(kk+2)]

### A convinience function to perform a single update single update
def one_step():
    x=X[i,:]
    err[0].send(np.dot(x,x))
    L[0][i]=err[0].get_sqrt()
    r1=Av.send(x)
    err[1].send(np.dot(r1,r1))
    L[1][i]=err[1].get_sqrt()/L[0][i]
    r2=r1
    for j in range(kk):
        r2=Eig[j].send(r2)
        err[j+2].send(np.dot(r2,r2))
        L[j+2][i]=err[j+2].get_sqrt()/L[0][i]
```

```python
f, ax = plt.subplots()
i=0

n_color=20
color_idx = np.linspace(0, 1, n_color) # for color cycling

for anim in range(100):
    for j in range(1):
        one_step() 
        i+=1;

    #print i
    #plt.figure(num=None, figsize=(5, 5), dpi=80, facecolor='w', edgecolor='k')
    ax.scatter(X[:,0],X[:,1])
    ax.scatter(X[i,0],X[i,1],c='r')
    Mean=Av.get_state()
    #print Mean,shift

    ax.scatter(Mean[0], Mean[1],c='g',s=100)
    for j in range(2):
        V=2*(Eig[j].get_state())
        #print V
        ax.arrow(Mean[0],Mean[1], V[0], V[1], head_width=1, head_length=1,ec=plt.cm.cool(color_idx[i % n_color]))
    ax.grid()
    ax.set_title(label='iteration '+str(i))
    plt.xlim([-10,10])
    plt.ylim([-10,10])
    #sleep(0.1)
    clear_output(wait=True)
    display(f)
    
plt.close()
```
